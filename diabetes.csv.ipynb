```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.metrics import accuracy_score
```

```python
diabetes=pd.read_csv("diabetes.csv")
```

```python
y=df.Outcome
x=df.drop('Outcome',axis=1)
```

```python
y.columns=["Outcome"]
```

```python

kmeans=KMeans(n_clusters=5)
clusters=kmeans.fit_predict(x)
print(len(clusters))
print(clusters)
```

```python
from scipy.stats import mode
labels=np.zeros_like(clusters)
print(len(labels))
print(clusters)
```

```python
for i in range(5):
    cat=(clusters==i)
    print(len(cat))
    print(cat)
```

```python
for i in range(4):
    cat=(clusters==i)
    labels[cat]=mode(y[cat])[0]
acc=accuracy_score(y,labels)
print('accuracy',acc)
```

```python
plt.figure(figsize=(10,10))
colormap=np.array(['red','lime','blue','Black'])
plt.subplot(2,2,1)
plt.scatter(x.Glucose,x.BloodPressure,c=colormap[y],s=40)
plt.title('real clusters')
plt.xlabel('Glucose')
plt.ylabel('BloodPressure')



plt.subplot(2,2,2)
plt.scatter(x.Glucose,x.BloodPressure,c=colormap[labels],s=40)
plt.title('kmeans clusters')
plt.xlabel('Glucose')
plt.ylabel('BloodPressure')
plt.show()
```
